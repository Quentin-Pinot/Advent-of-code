# Part 2

```python
DAY = 3
PART = 2
s = get_input(DAY).strip()
```

```python
tab = []

for i in s.split('\n'):
    print(i)
    game_match = re.match(r"Game (\d+): (.+)", i)
    bag = {"red": 0, "green": 0, "blue": 0}

    for cube in game_match.group(2).split(';'):
        for v, k in re.findall(r"(\d+) (\w+)", cube):
            print(k, v)
            bag[k] = max(bag[k], int(v))

    tab.append(reduce(lambda x, y: x * y, bag.values(), 1))
```

```python
sum(tab)
```

```python
submit(DAY, PART, sum(tab))
```
