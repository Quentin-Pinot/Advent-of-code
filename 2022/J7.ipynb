```python
class file:
    def __init__(self, size, name):
        self.size = size
        self.name = name

class dir:
    def __init__(self, name, path):
        self.name = name
        self.path = path
        self.content = {}
        self.size = 0

    def addContent(self, size_or_Parent, name):
        if size_or_Parent == 'dir' : 
            self.content[name] = dir(name, size_or_Parent)
        else:
            self.content[name] = file(int(size_or_Parent), name)
```
